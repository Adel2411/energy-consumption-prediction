```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the data from the CSV files
file1 = pd.read_csv('data/train_energy.csv')
file2 = pd.read_csv('result/submission32_best.csv')
# file2 = pd.read_csv('result/submission31_best2.csv')

# Trying the current generated forecast
# file2 = pd.read_csv('skforecast_submission.csv')

# Trying best moh's forecast
# file2 = pd.read_csv('moh/to_adel.csv')
```

```python

# Convert the 'Date' column to datetime type
file1['DateTime'] = pd.to_datetime(file1['DateTime'])
file2['Date'] = pd.to_datetime(file2['DateTime'])

print(file1.head())
print(file2.head())
```

```python

# Ensure the data is sorted by date
file1 = file1.sort_values(by='DateTime')
file2 = file2.sort_values(by='DateTime')

# Concatenate the data frames
combined_data = pd.concat([file1, file2])
```

```python

# Plot the data
plt.figure(figsize=(15, 6))
plt.plot(combined_data['DateTime'], combined_data['EnergyConsumption(kWh)'], label='Energy Consumption')
plt.axvline(x=file1['DateTime'].max(), color='red', linestyle='--', label='Transition Point')
plt.title('Energy Consumption Over Time')
plt.xlabel('DateTime')
plt.ylabel('Energy Consumption (kWh)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
