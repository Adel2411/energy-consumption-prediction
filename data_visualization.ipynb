# SOAI Lab: Smart Home Energy Consumption Prediction


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the training data
train_data = pd.read_csv('data/train_energy.csv')

# Convert DateTime column to datetime type
train_data['DateTime'] = pd.to_datetime(train_data['DateTime'])

# Rename columns to 'ds' for DateTime and 'y' for the target variable
train_data.rename(columns={'DateTime': 'ds', 'EnergyConsumption(kWh)': 'y'}, inplace=True)
```

```python
# Plot histogram of energy consumption
plt.figure(figsize=(10, 6))
plt.hist(train_data['y'], bins=50, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Energy Consumption', fontsize=16)
plt.xlabel('Energy Consumption (kWh)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Plot boxplot of energy consumption
plt.figure(figsize=(8, 5))
sns.boxplot(x=train_data['y'], color='skyblue')
plt.title('Boxplot of Energy Consumption', fontsize=16)
plt.xlabel('Energy Consumption (kWh)', fontsize=14)
plt.show()
```

```python
# Plot distribution of energy consumption by hour
plt.figure(figsize=(12, 6))
train_data['hour'] = train_data['ds'].dt.hour
sns.boxplot(x='hour', y='y', data=train_data, palette='viridis')
plt.title('Energy Consumption Distribution by Hour of Day', fontsize=16)
plt.xlabel('Hour of Day', fontsize=14)
plt.ylabel('Energy Consumption (kWh)', fontsize=14)
plt.show()
```

```python
# Plot distribution of energy consumption by day of the week
plt.figure(figsize=(12, 6))
train_data['day_of_week'] = train_data['ds'].dt.dayofweek
sns.boxplot(x='day_of_week', y='y', data=train_data, palette='viridis')
plt.title('Energy Consumption Distribution by Day of Week', fontsize=16)
plt.xlabel('Day of Week (0=Monday, 6=Sunday)', fontsize=14)
plt.ylabel('Energy Consumption (kWh)', fontsize=14)
plt.show()
```

```python
# Plot data of one day
one_day = train_data[train_data['ds'].dt.date == pd.Timestamp('2016-01-01').date()]

plt.figure(figsize=(12, 6))
plt.plot(one_day['ds'], one_day['y'], marker='o', linestyle='-', color='blue')
plt.title('Energy Consumption on 2016-01-01', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Energy Consumption (kWh)', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Plot data of one week
one_week = train_data[(train_data['ds'] >= '2016-01-01') & (train_data['ds'] < '2016-01-08')]

plt.figure(figsize=(12, 6))
plt.plot(one_week['ds'], one_week['y'], marker='o', linestyle='-', color='green')
plt.title('Energy Consumption from 2016-01-01 to 2016-01-07', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Energy Consumption (kWh)', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Plot data of one month
one_month = train_data[(train_data['ds'] >= '2016-12-01') & (train_data['ds'] < '2017-01-01')]

plt.figure(figsize=(12, 6))
plt.plot(one_month['ds'], one_month['y'], marker='o', linestyle='-', color='purple')
plt.title('Energy Consumption in January 2016', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Energy Consumption (kWh)', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Plot data of one year
one_year = train_data[train_data['ds'].dt.year == 2016]

plt.figure(figsize=(12, 6))
plt.plot(one_year['ds'], one_year['y'], marker='o', linestyle='-', color='red')
plt.title('Energy Consumption in 2016', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Energy Consumption (kWh)', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```
